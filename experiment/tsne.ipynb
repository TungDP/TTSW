```python
import sys
sys.path.append('../src')
from utilities import run_tsne
```

```python
# ===== CẤU HÌNH CHUNG =====
alg          = "TTSWD"       # thuật toán
datatype     = "UCR_TSL"     # định dạng dữ liệu
num_runs     = 5             # số lần chạy (seed = run_idx * 4)
n_components = 2             # số chiều embedding
result_dir   = "../result"   # thư mục lưu: result/<alg>_tsne.xlsx + result/<dataset>/<alg>_tsne_seed0.png

# ===== THAM SỐ THUẬT TOÁN =====
# Chỉ cần khai báo các key muốn thay đổi so với mặc định.
# TTSWD defaults: n_trees=32, time_weight=64.0, distance_mode="auto",
#                 streaming_block_size=2048, distance_scale=1.0
alg_kwargs = {
    "n_trees": 64,
    "time_weight": 1,
    "distance_mode": "streaming",
    # "streaming_block_size": 2048,
    "distance_scale": 1.0,
}

# ===== DATASETS =====
# Để trống skip_datasets = set() để chạy tất cả.
skip_datasets = set()

all_datasets = [
    "ArrowHead",               # AH
    "BasicMotions",            # BM
    "BeetleFly",               # BF
    "CBF",                     # CBF
    "Chinatown",               # CT
    "CinCECGTorso",            # CET
    "DiatomSizeReduction",     # DSR
    "GunPointAgeSpan",         # GPA
    "GunPointMaleVersusFemale",# GPM
    "GunPointOldVersusYoung",  # GPO
    "Ham",                     # Ham
    "InsectEPGRegularTrain",   # IERT
    "ItalyPowerDemand",        # IPD
    "Meat",                    # Meat
    "MelbournePedestrian",     # MP
    "MixedShapesSmallTrain",   # MS2T
    "MoteStrain",              # MS
    "OliveOil",                # O2
    "Plane",                   # Plane
    "SmoothSubspace",          # S2
]
```

```python
datasets_to_run = [d for d in all_datasets if d not in skip_datasets]
run_tsne(
    datasets_to_run,
    datatype,
    alg,
    num_runs=num_runs,
    n_components=n_components,
    result_dir=result_dir,
    alg_kwargs=alg_kwargs,
)
```
