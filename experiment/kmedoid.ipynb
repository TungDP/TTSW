```python
import sys
sys.path.append('../src')
from utilities import run_kmedoid
```

```python
# ===== CẤU HÌNH CHUNG =====
alg        = "TTSWD"        # thuật toán
datatype   = "UCR_TSL"      # định dạng dữ liệu
n_clusters = None            # None = dùng số lớp thực tế của dataset
num_runs   = 5              # số lần chạy (seed = run_idx)
result_dir = "../result"    # thư mục lưu kết quả: result/<alg>_kmedoid.xlsx

# ===== THAM SỐ THUẬT TOÁN =====
# Chỉ cần khai báo các key muốn thay đổi so với mặc định.
# TTSWD defaults: n_trees=32, time_weight=64.0, distance_mode="auto",
#                 streaming_block_size=2048, distance_scale=1.0
alg_kwargs = {
    "n_trees": 64,
    "time_weight": 10.0,
    "distance_mode": "auto",
    "streaming_block_size": 2048,
    "distance_scale": 1.0,
}

all_datasets = [
    "ArrowHead",               # AH
    "BasicMotions",            # BM
    "BeetleFly",               # BF
    "CBF",                     # CBF
    "Chinatown",               # CT
    "CinCECGTorso",            # CET
    "DiatomSizeReduction",     # DSR
    "GunPointAgeSpan",         # GPA
    "GunPointMaleVersusFemale",# GPM
    "GunPointOldVersusYoung",  # GPO
    "Ham",                     # Ham
    "InsectEPGRegularTrain",   # IERT
    "ItalyPowerDemand",        # IPD
    "Meat",                    # Meat
    "MelbournePedestrian",     # MP
    "MixedShapesSmallTrain",   # MS2T
    "MoteStrain",              # MS
    "OliveOil",                # O2
    "Plane",                   # Plane
    "SmoothSubspace",          # S2
]

# Bỏ qua các dataset (để trống = chạy tất cả):
skip_datasets = set()
```

```python
datasets_to_run = [d for d in all_datasets if d not in skip_datasets]
run_kmedoid(
    datasets_to_run,
    datatype,
    alg,
    n_clusters=n_clusters,
    num_runs=num_runs,
    result_dir=result_dir,
    alg_kwargs=alg_kwargs,
)
```
