```python
import sys
sys.path.append('../src')
from utilities import run_knn
```

```python
# ===== CẤU HÌNH CHUNG =====
alg               = "TTSWD"        # thuật toán
datatype          = "UCR_TSL"      # định dạng dữ liệu
num_neighbor_list = [1]            # danh sách k cho kNN
num_runs          = 5              # số lần chạy (seed = run_idx * 4)
result_dir        = "../result"    # thư mục lưu kết quả: result/<alg>_knn.xlsx

# ===== THAM SỐ THUẬT TOÁN =====
# Chỉ cần khai báo các key muốn thay đổi so với mặc định.
# TTSWD defaults: n_trees=112, time_weight=64.0, distance_mode="auto",
#                 streaming_block_size=8192, distance_scale=1.0
alg_kwargs = {
    "n_trees": 64,
    "time_weight": 10,
    "distance_mode": "streaming",
    "streaming_block_size": 1024,
    "distance_scale": 1.0,
    "approx_diameter": True,
}

# ===== DATASETS =====
# Để trống skip_datasets = set() để chạy tất cả.
skip_datasets = set()

all_datasets = [
    "ArrowHead",               # AH
    "BasicMotions",            # BM
    "BeetleFly",               # BF
    "CBF",                     # CBF
    "Chinatown",               # CT
    "CinCECGTorso",            # CET
    "DiatomSizeReduction",     # DSR
    "GunPointAgeSpan",         # GPA
    "GunPointMaleVersusFemale",# GPM
    "GunPointOldVersusYoung",  # GPO
    "Ham",                     # Ham
    "InsectEPGRegularTrain",   # IERT
    "ItalyPowerDemand",        # IPD
    "Meat",                    # Meat
    "MelbournePedestrian",     # MP
    "MixedShapesSmallTrain",   # MS2T
    "MoteStrain",              # MS
    "OliveOil",                # O2
    "Plane",                   # Plane
    "SmoothSubspace",          # S2
]
```

```python
datasets_to_run = [d for d in all_datasets if d not in skip_datasets]
run_knn(
    datasets_to_run,
    datatype,
    alg,
    num_neighbor_list=num_neighbor_list,
    num_runs=num_runs,
    result_dir=result_dir,
    alg_kwargs=alg_kwargs,
)
```
